```python
import os
import shutil

def move_files(src_path, move_path):
    # Pastikan direktori tujuan ada, jika tidak, buat
    if not os.path.exists(move_path):
        os.makedirs(move_path)
    
    # Telusuri file di direktori sumber
    for root, dirs, files in os.walk(src_path):
        for file in files:
            # Cek apakah file mengandung "closed_eye" di namanya
            if "closed_eye" not in file:
                # Pindahkan file ke direktori tujuan
                src_file_path = os.path.join(root, file)
                dst_file_path = os.path.join(move_path, file)
                shutil.move(src_file_path, dst_file_path)
                print(f"Moved: {src_file_path} -> {dst_file_path}")
            else:
                print(f"Kept: {os.path.join(root, file)}")

# Contoh penggunaan
src_path = './dataset/Eksternal/Mata/Tertutup'
move_path = './dataset/Eksternal/Mata/Terbuka'
move_files(src_path, move_path)
```

```python
def count_files_in_directory(directory):
    file_count = 0
    for root, dirs, files in os.walk(directory):
        file_count += len(files)
    
    return file_count


print('Open Eye: ', count_files_in_directory(f'testing/Mata/Terbuka'))
print('Close Eye: ', count_files_in_directory(f'testing/Mata/Tertutup'))
```

```python
import os

def rename_files_sequentially(directory):
    # List untuk menyimpan path semua file yang ditemukan
    file_paths = []

    # Menelusuri directory dan subdirectory untuk mengumpulkan semua file
    for root, dirs, files in os.walk(directory):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            file_paths.append(file_path)

    # Mengurutkan file berdasarkan path
    file_paths.sort()

    # Mengganti nama file secara sekuensial
    for i, file_path in enumerate(file_paths):
        # Mengambil direktori dan ekstensi file
        dir_name = os.path.dirname(file_path)
        ext = os.path.splitext(file_path)[1]

        # Membuat nama file baru
        new_name = f"{i+1:04d}{ext}"  # 4 digit padding (misal: 0001.ext, 0002.ext, ...)

        # Membuat path baru untuk file yang sudah diubah namanya
        new_file_path = os.path.join(dir_name, new_name)

        # Mengganti nama file
        os.rename(file_path, new_file_path)

        print(f"Renamed: {file_path} -> {new_file_path}")

# Contoh penggunaan
input_directory = './testing'
rename_files_sequentially(input_directory)
```
